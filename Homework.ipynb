#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 2
num2 = 5.4

result = num1 + num2
print(result)

# Shorthand - Version
result += 2 # result = result + 2
print(result)
```

##### Subtraction

```python
result_diff = num2 - num1
print(result_diff)

#shorthand - version
result_diff -= 1 # result_diff = result_diff - 1
print(result_diff)
```

##### Multiplication

```python
result_mul = num1 * num2
print(result_mul)

result_mul *= 2
print(result_mul)
```

##### Division

```python
result_div = num2 / num1
print(result_div)

result_div /= 2
print(result_div)
```

##### Floor Division

```python
result_floor = num2 // num1
print(result_floor)

# shorthand - version
result_floor //= 2
print(result_floor)
```

##### Modulo

```python
result_mod = num2 % num1
print(result_mod)

#shorthand - version
result_mod %= 2
print(result_mod)
```

##### Exponential

```python
square = 5**2
print(square)

#shorthand 
square **=2
print(square)
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 5
num2 = 10

if num2 == num1: 
    print('Equal Values')
else:
    print('Not Equal')
    
if num2 >= num1:
    print('Num2 is greater')
elif num1 <= num2:
    print("Num1 is less than")
```

##### 'is' keyword

## Using 'and'/'or' with If Statements

```python
# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num_4 = 15
num_5 = 3
num_6 = 10
num_7 = 3

if num_4 / 5 == num_5 and num_6 - 7 == num_7:
    print('True AND True')
    
if num_4 > num_5 or num_6 == num_7:
    print('True OR False')
```

##### Indexing a List

```python
# [start: stop: step] 

# Single Index
print(names[0])

# Print starting at index 1 going to the end
print(names[1:])

# Print starting at beginning until n is reached (n == number we want to stop with)
print(names[:2])

# Print starting at index 1 going up by 2 in each iteration
print(names[1::2])

# Print starting from back, reverse order
print(names[::-1])
```

```python
# HINT::
# An else after an if runs if the if didnt
# An else after a for runs if the for didnt break
for x in range(100):
    if x%2 != 0 and x%3 != 0 and x%5 != 0 and x%7 != 0:
        print(x)
    if x == 2 or x == 3 or x == 5 or x == 7:
        print(x)
```
